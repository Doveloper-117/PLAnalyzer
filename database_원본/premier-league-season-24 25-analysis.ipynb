# üèÅ 1. Importing Libraries

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Optional: For interactive charts
import plotly.express as px
```

# üìä 2. Basic Data Overview

```python
df=pd.read_csv('/kaggle/input/english-premier-league-player-stats-2425/epl_player_stats_24_25.csv')
```

```python

print("Shape of dataset:", df.shape)
print("Columns:", df.columns.tolist())
df.info()
df.describe()
```

```python
# üßπ 4. Data Cleaning (Optional)
# Convert Minutes to numeric (if needed)
df['Minutes'] = pd.to_numeric(df['Minutes'], errors='coerce')

# Fill NaNs in numeric columns with 0
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(0)
```

# 3. Premier League 2024 Player Stats: Performance Analysis

```python
# Example: Top 10 goal scorers
top_scorers = df.sort_values(by="Goals", ascending=False).head(10)
print(top_scorers)
# Example: Filter for Goalkeepers
goalkeepers = df[df["Position"] == "GKP"]
print(goalkeepers)
# Example: Club-wise average goals
club_avg_goals = df.groupby("Club")["Goals"].mean().sort_values(ascending=False)
print(club_avg_goals)
```

```python
# üßÆ 5. Feature Engineering
df['Goals_per90'] = (df['Goals'] / df['Minutes']) * 90
df['Assists_per90'] = (df['Assists'] / df['Minutes']) * 90
df['Shot_Accuracy'] = np.where(df['Shots'] > 0, (df['Shots On Target'] / df['Shots']) * 100, 0)
df['Discipline_Ratio'] = (df['Yellow Cards'] + df['Red Cards']) / df['Appearances']

df[['Player Name', 'Position', 'Goals_per90', 'Assists_per90', 'Shot_Accuracy']].head()
```

```python
# üßç‚Äç‚ôÇÔ∏è 6. Top Performers by Position
top_forwards = df[df['Position'] == 'FWD'].sort_values('Goals_per90', ascending=False).head(10)
top_mids = df[df['Position'] == 'MID'].sort_values('Assists_per90', ascending=False).head(10)
top_gks = df[df['Position'] == 'GKP'].sort_values('Saves %', ascending=False).head(10)

```

```python
# üìà 7. Visualizations

# Top 10 Goal Scorers
top_scorers = df.sort_values('Goals', ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='Goals', y='Player Name', data=top_scorers, palette='viridis')
plt.title("Top 10 Goal Scorers")
plt.xlabel("Goals")
plt.ylabel("Player")
plt.show()
```

```python
# ‚öôÔ∏è 8. Optional Clustering (Advanced)
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

features = df[['Goals_per90', 'Assists_per90', 'Shot_Accuracy']].fillna(0)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)

kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

px.scatter_3d(df, x='Goals_per90', y='Assists_per90', z='Shot_Accuracy',
              color='Cluster', hover_name='Player Name', title='Player Clusters')
```

# 4. Shot Efficiency Analysis: Who Wastes Chances?

```python
df['Shot_Accuracy'] = np.where(df['Shots'] > 0, (df['Shots On Target'] / df['Shots']) * 100, 0)
df['Conversion_Rate'] = np.where(df['Shots'] > 0, (df['Goals'] / df['Shots']) * 100, 0)

# Top 10 Efficient Finishers (Min 20 shots)
efficient = df[df['Shots'] > 20].sort_values('Conversion_Rate', ascending=False).head(10)
efficient[['Player Name', 'Club', 'Shots', 'Goals', 'Conversion_Rate']]
```

```python

sns.scatterplot(data=efficient, x='Shot_Accuracy', y='Conversion_Rate', hue='Club')
plt.title("Efficiency vs Accuracy (Top Finishers)")
```

# 5. Build Your Team: Best XI by Stats

```python
# Strip column names just in case
df.columns = df.columns.str.strip()

# Ensure numeric
df['Appearances'] = pd.to_numeric(df['Appearances'], errors='coerce')
df['Minutes'] = pd.to_numeric(df['Minutes'], errors='coerce')

# New combined filter
df_filtered = df[(df['Appearances'] >= 10) & (df['Minutes'] >= 500)].copy()
```

```python


best_xi = {}

# üß§ Goalkeeper with highest save % (min 10 appearances)
gkp_df = df_filtered[df_filtered['Position'] == 'GKP']
best_xi['GKP'] = gkp_df.sort_values('Saves %', ascending=False).head(1)

# üõ°Ô∏è Defenders - 4 with highest goals, then lower minutes (if tie)
def_df = df_filtered[df_filtered['Position'] == 'DEF']
defenders = def_df.sort_values(['Goals', 'Minutes'], ascending=[False, True]).head(4)

# üß† Midfielders - 3 with highest goal contribution
mid_df = df_filtered[df_filtered['Position'] == 'MID'].copy()
mid_df['Goal_Contrib'] = mid_df['Goals'] + mid_df['Assists']
midfielders = mid_df.sort_values('Goal_Contrib', ascending=False).head(3)

# üéØ Forwards - 3 with highest goals per 90 mins
fwd_df = df_filtered[df_filtered['Position'] == 'FWD'].copy()
fwd_df['Goals_per90'] = (fwd_df['Goals'] / fwd_df['Minutes']) * 90
fwd_df['Goals_per90'] = fwd_df['Goals_per90'].replace([np.inf, -np.inf], 0).fillna(0)
forwards = fwd_df.sort_values('Goals_per90', ascending=False).head(3)

# ‚úÖ Combine all players
best_xi_df = pd.concat([best_xi['GKP'], defenders, midfielders, forwards], ignore_index=True)

# üìä Display final Best XI
best_xi_df[['Player Name', 'Position', 'Club', 'Goals', 'Assists', 'Minutes']]
```

# 6. Discipline Watch: Who‚Äôs the Most Aggressive?

```python
df['Total_Cards'] = df['Yellow Cards'] + df['Red Cards']
df['Fouls_per_Game'] = df['Fouls'] / df['Appearances']

# Most Aggressive
aggro = df.sort_values(['Total_Cards', 'Fouls_per_Game'], ascending=False).head(10)
aggro[['Player Name', 'Club', 'Position', 'Total_Cards', 'Fouls_per_Game']]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Total_Cards', y='Player Name', data=aggro, palette='Reds_r')
plt.title("Top 10 Most Booked Players")
```

# 7. Minutes vs Impact: Who Deserves More Play Time?

```python
underplayed = df[(df['Minutes'] < 1000) & (df['Appearances'] >= 5)].copy()
underplayed['Goals_per90'] = (underplayed['Goals'] / underplayed['Minutes']) * 90
underplayed['Assists_per90'] = (underplayed['Assists'] / underplayed['Minutes']) * 90
underplayed['Impact'] = underplayed['Goals_per90'] + underplayed['Assists_per90']

# Top 10 impact players with limited minutes
impact_players = underplayed.sort_values('Impact', ascending=False).head(10)
impact_players[['Player Name', 'Minutes', 'Goals', 'Assists', 'Goals_per90', 'Impact']]
```

# 8. Clutch Players: Who Performs in Big Minutes?

```python
high_minutes = df[df['Minutes'] > 2000].copy()
high_minutes['Goals_per90'] = (high_minutes['Goals'] / high_minutes['Minutes']) * 90
high_minutes['Assists_per90'] = (high_minutes['Assists'] / high_minutes['Minutes']) * 90

top_clutch = high_minutes.sort_values(by='Goals_per90', ascending=False).head(10)
top_clutch[['Player Name', 'Minutes', 'Goals', 'Goals_per90']]
```

# 9. Club Dependency: Who Carries the Team?

```python
# Make a copy to avoid modifying the original df
df_copy = df.copy()

# Group by Club to compute total goals and assists per team
club_stats = df_copy.groupby('Club')[['Goals', 'Assists']].sum().reset_index()
club_stats.columns = ['Club', 'Team_Goals', 'Team_Assists']

# Merge team totals back into player-level DataFrame
df_copy = df_copy.merge(club_stats, on='Club')

# Calculate each player's goal contribution percentage
df_copy['Goal_Contribution_Pct'] = ((df_copy['Goals'] + df_copy['Assists']) / 
                                    (df_copy['Team_Goals'] + df_copy['Team_Assists'])) * 100

# Filter players with decent playtime and sort by contribution %
key_players = df_copy[df_copy['Minutes'] > 900].copy()
key_players = key_players.sort_values('Goal_Contribution_Pct', ascending=False).head(10)

# Show key players
key_players_display = key_players[['Player Name', 'Club', 'Goals', 'Assists', 'Goal_Contribution_Pct']]
display(key_players_display)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=key_players, x='Goal_Contribution_Pct', y='Player Name', hue='Club', dodge=False, palette='viridis')
plt.title('Top 10 Most Essential Players to Their Clubs')
plt.xlabel('Goal Contribution (%)')
plt.ylabel('Player')
plt.xlim(0, 100)
plt.grid(axis='x')
plt.show()
```
